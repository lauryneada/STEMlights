```python
#Import Libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Get url, parse & print it
url='http://createsend.com/t/t-166AD5026249E8D12540EF23F30FEDED'
r=requests.get(url)

article = BeautifulSoup(r.text, 'html.parser')
print(article.prettify())
```

```python
#Export resulting html file
with open ("articletest.html","w") as file:
    file.write(str(article))
```
